```python
# универсальный код, который переводит torch тензор на ту платформу, 
# которая у нас сейчас в компьютере есть
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
device
```

```python
x_cuda = x.to(device) # переводит тензор на gpu / cpu
x_cuda
```

```python
# %time - сколько времени выполняется код
%time y = (x - x + x * 10.0) ** 2
```

```python
%time y_cuda = (x_cuda - x_cuda + x_cuda * 10.0) ** 2
```
